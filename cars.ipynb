1. Can you find that there is improvement in mpg over the years ?


```python
import seaborn as sns

import matplotlib.pyplot as plt

# Load the dataset
mpg = sns.load_dataset('mpg')

# Check for improvement in mpg over the years
plt.figure(figsize=(10, 6))
sns.lineplot(x='model_year', y='mpg', data=mpg, ci=None)
plt.title('MPG Improvement Over the Years')
plt.xlabel('Model Year')
plt.ylabel('Miles per Gallon (MPG)')
plt.show()
```

2.Does cubicinches of a vehicle affects the vehicles time to reach 60 Miles per hour?

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='displacement', y='acceleration', data=mpg, scatter_kws={'s':10}, line_kws={'color':'red'})
plt.title('Displacement vs Acceleration')
plt.xlabel('Displacement (cubic inches)')
plt.ylabel('Acceleration (0-60 mph in seconds)')
plt.show()
```

3.Does a specific country prefer vehicles with more horse power ?


```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='origin', y='horsepower', data=mpg)
plt.title('Horsepower Distribution by Country of Origin')
plt.xlabel('Country of Origin')
plt.ylabel('Horsepower')
plt.show()
```

4.How does the horse power affects the average of the vehicle ?

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='horsepower', y='mpg', data=mpg, scatter_kws={'s':10}, line_kws={'color':'blue'})
plt.title('Horsepower vs MPG')
plt.xlabel('Horsepower')
plt.ylabel('Miles per Gallon (MPG)')
plt.show()
```

5.How does the cubic inches affects the average of the vehicle ?

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='displacement', y='mpg', data=mpg, scatter_kws={'s':10}, line_kws={'color':'green'})
plt.title('Displacement vs MPG')
plt.xlabel('Displacement (cubic inches)')
plt.ylabel('Miles per Gallon (MPG)')
plt.show()
```

6.What is the effect of number of cylinders on the mpg ?

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='cylinders', y='mpg', data=mpg)
plt.title('Effect of Number of Cylinders on MPG')
plt.xlabel('Number of Cylinders')
plt.ylabel('Miles per Gallon (MPG)')
plt.show()
```

7.What relation between cubic inches , number of cylinders and horse power you can observe from the data ?

```python
sns.pairplot(mpg, vars=['displacement', 'cylinders', 'horsepower'], kind='reg', diag_kind='kde')
plt.suptitle('Pair Plot of Displacement, Cylinders, and Horsepower', y=1.02)
plt.show()
```

8.How much proportion of the total records is occupied by vehicles with 8 cylinders ?

```python
# Calculate the proportion of vehicles with 8 cylinders
proportion_8_cylinders = (mpg['cylinders'] == 8).mean()
print(f"Proportion of vehicles with 8 cylinders: {proportion_8_cylinders:.2%}")
```
